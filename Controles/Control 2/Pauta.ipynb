**11.** El valor promedio de venta de cada tienda en marzo (con sus respectivos nombres o ids). **(0.5 pts)**

- 0.1 por intervalo de fechas correcta (marzo 2021)

**OJO con los group by**

```python
%%sql
SELECT Stores.name, AVG(bill_sales.sales) AS average
FROM Stores JOIN (
    SELECT Stores.id AS store_id, SUM(BillVariants.quantity * BillVariants.price * ((100 - BillVariants.discount_rate)*1.0/100)) AS sales
    FROM Stores, Bills, Products, Variants, BillVariants
    WHERE Stores.id = Products.store_id
    AND Variants.product_id = Products.id
    AND BillVariants.variant_id = Variants.id
    AND BillVariants.bill_id = Bills.id
    AND Bills.created_at BETWEEN '2021-03-01' AND '2021-04-01'
    GROUP BY Stores.id, Bills.id) AS bill_sales
ON Stores.id = bill_sales.store_id
GROUP BY Stores.id
```

**12.** El porcentaje del total de clientes que fueron recurrentes en el mes de abril (recurrentes son aquellos que ya habían comprado alguna vez antes de comprar en abril). Los clientes se identifican por el email **(0.5 pts)**

Pauta: 
- 0.3 por calcular bien las recurrencia
- 0.2 por el resto

```python
%%sql
SELECT COUNT(DISTINCT old_clients.email)*1.0/COUNT(DISTINCT Bills.email) * 100 AS recurrence
FROM Bills LEFT OUTER JOIN (
    SELECT Bills.email AS email
    FROM Bills
    WHERE Bills.created_at < '2021-04-01'
    ) AS old_clients
ON Bills.email = old_clients.email
WHERE Bills.created_at > '2021-04-01'
```

**13.**  La cantidad de dias en promedio que han pasado desde que se creo cada venta, agrupado por status. **(0.5 pts)**

**Pro Tips**:
- `DATE(SUBSTR(columna, 0, 11))` Retorna un Date de sqlite a partir de una columna de fecha de nuesta base de datos.
-`DATE('now')` Retorna un objeto Date de sqlite con la fecha de hoy
-La función [JULIANDAY(date)](https://www.techonthenet.com/sqlite/functions/julianday.php#:~:text=The%20SQLite%20julianday%20function%20takes,as%20a%20floating%20point%20number.) Retorna los días que han pasado entre date y una fecha constante muy en el pasado.

Pauta: 
- 0.4 por calcular bien la cantidad de días promedio
- 0.1 agrupar por status

```python
%%sql
SELECT Bills.status, AVG(JULIANDAY(DATE('now')) - JULIANDAY(DATE(SUBSTR(Bills.created_at, 0, 11)))) AS average_diff
FROM Bills
GROUP BY Bills.status
```

## Parte 2: Índices (1 punto, 0.1 cada respuesta) 

En esta parte del control tienen qué responder cual será el costo I/O para ejecutar consultas SQL ocupando distintos tipos de índices.

En particular, en este ejercicio asumiremos qué tenemos una relación Alumno(id int PRIMARY KEY, nombre varchar(100), promedio float).

A nuestra disposición encontramos dos índices: 
- Un B+tree *clustered* sobre id, y 
- Un hash index *unclustered* sobre el promedio.

La relación tiene 320,000 tuplas. En cada página del disco nos cabe 100 tuplas.El promedio toma valores entre 3.9 y 7.0, en incrementos de 0.1 (i.e. hay solo 32 posibles promedio). Los ids van desde 1 a 320,000.

El árbol es de altura *h*, y la hojas están ocupadas a 80%. 

En el Hash Index cada página contiene 1,000 punteros a las páginas del disco, y hay 32 buckets (uno para cada promedio). Las notas son distribuidas de manera uniforme y **si hay overflow pages** (deberían calcular cuanto).

¿Cuál es el costo de evaluar las siguientes consultas con los dos índices (en el peor de los casos)?
1. SELECT * FROM Alumno WHERE id=12
2. SELECT * FROM Alumno WHERE nombre = 'Adrián Soto'
3. SELECT * FROM Alumno WHERE promedio = 5.6
4. SELECT * FROM Alumno WHERE id > 1000 AND id <= 2000
5. SELECT * FROM Alumno WHERE promedio <4.9

## Solución

En una página del disco caben 100 tuplas.

La relación entera tiene 3,200 páginas en el disco.

Cada bucket del Hash Index tiene 9 overflow pages (y una principal). En total hay 32 x 10 x 1000 = 320,000 tuplas.

| Consulta| B+ tree | Hash Index |
| ------- | -------- | ------------ |
| 1       | h | 320 (índice) + 320.000 (si lo hacemos a la mala), o 320 + 3,200 (si chequeamos si ya tuvimos la página) |
| 2       | (h - 1) + 3,200/0.8 | 320 (índice) + 320.000 |
| 3    | (h - 1) + 3,200/0.8 | 10 + 10.000 |
| 4    | h + 10/0.8 | 320 (índice) + 320.000 |
| 5    | (h - 1) + 3,200/0.8 | 10 * 10 + 10 * 10 * 1,000|
